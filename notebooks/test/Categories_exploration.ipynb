# Categories selection

```python
import pandas as pd 
```

```python
# Uploading the table of categories available in https://huggingface.co/datasets/McAuley-Lab/Amazon-Reviews-2023
df_categories = pd.read_csv("data/Category_groups.csv", index_col="Unnamed: 0")
df_categories.head()
```

```python
# Converting str into int
def convert_to_int(s):
    # Check if the string ends with 'K' or 'M' and convert accordingly
    if s.endswith("K"):
        return int(float(s[:-1]) * 1_000)
    elif s.endswith("M"):
        return int(float(s[:-1]) * 1_000_000)
    elif s.endswith("B"):
        return int(float(s[:-1]) * 1_000_000_000)
    else:
        # If there's no suffix, just convert directly to int
        return int(s)
df_categories["User"] = df_categories["User"].apply(convert_to_int)
df_categories["item"] = df_categories["item"].apply(convert_to_int)
df_categories["Rating"] = df_categories["Rating"].apply(convert_to_int)
df_categories["R_token"] = df_categories["R_token"].apply(convert_to_int)
df_categories["M_token"] = df_categories["M_token"].apply(convert_to_int)
```

```python
# In order to assure a good quality of data (especially textual data), we will try to evaluate
# each category based on the following critera
#  - number of tokens per reviews
#  - number of reviews per product 

df_categories["tokens_per_reviews"] = df_categories.R_token / df_categories.Rating
df_categories["reviews_per_product"] = df_categories.Rating / df_categories.item
```
